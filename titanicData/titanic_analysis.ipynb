```python
def plot_variable_distribution(variable):
    cat=train_data[variable]
    counts=cat.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(counts.index,counts)
    plt.xticks(counts.index,counts.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print(variable,counts)
```

```python
category1=["Survived","Sex","Pclass","Embarked","SibSp","Parch"]
```

```python
for i in category1:
    plot_variable_distribution(i)
```

```python
category2=["Cabin","Name","Ticket"]
for i in category2:
    print(train_data[i].value_counts())
```

```python
def hist(variable):
    plt.figure(figsize=(9,3))
    plt.hist(train_data[variable])
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
```

```python
numerical=["Fare","Age","PassengerId"]
for i in numerical:
    hist(i)
```
